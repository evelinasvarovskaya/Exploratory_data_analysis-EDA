О датасете: 

- набор данных представляет из себя даннные о погоде, дате, температуре, скорости ветра, видимости в этот день и тд. 

1. Импортируем и получим данные. Далее посмотрим основную информацию:
* head() — показывает первые N строк данных (по умолчанию N=5).
* shape — показывает общее количество строк и столбцов
* columns — показывает имя каждого столбца.
* dtypes — показывает тип данных каждого столбца.
* unique() — в столбце отображаются все уникальные значения. Его можно применять только к одному столбцу, а не ко всему кадру данных.
* nunique() — показывает общее количество. уникальных значений в каждом столбце. Его можно применять как к одному столбцу, так и ко всему кадру данных.
* count — показывает общее количество. ненулевых значений в каждом столбце. Его можно применять как к одному столбцу, так и ко всему кадру данных.
* value_counts — в столбце отображаются все уникальные значения с их количеством. Его можно применить только к одному столбцу.
* info() — предоставляет основную информацию о данных.
